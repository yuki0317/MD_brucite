```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack
import pandas as pd
```

```python
with open('vac_test.dat', 'r') as file:
    times, atoms1, atoms2, atoms3 = np.array([]), np.array([]), np.array([]), np.array([])
    for line in file:
        items=line.split(' ')
        times=np.append(times, float(items[0]))
        atoms1=np.append(atoms1, float(items[1]))
        atoms2=np.append(atoms2, float(items[2]))
        atoms3=np.append(atoms3, float(items[3]))
```

```python
times
```

```python
time_step = float(times[1] - times[0])
```

```python
time_step
```

```python
#FFT for atom1
sample_freq = fftpack.fftfreq(atoms1[:].size, d=time_step)
y_fft = fftpack.fft(atoms1[:])
pidxs = np.where(sample_freq > 0)
freqs, power = sample_freq[pidxs], np.abs(y_fft)[pidxs]
freq = freqs[power.argmax()]
```

```python
#Plot
plt.figure(figsize=(8,10))
plt.subplot(2, 1, 1)
#plt.plot(time, atoms)
plt.plot(times,atoms1,'b-', linewidth=1)
plt.xlabel('Time [ps]')
plt.ylabel('Amplitude')
plt.grid(True)
#
plt.subplot(2, 1, 2)
#plt.loglog(freqs, power)
plt.loglog(freqs, power,'b.-',lw=1)
plt.xlabel('Frequency')
plt.ylabel('Power')
plt.grid(True)

plt.show()
```
